
# Description of dataset

## Dataset: World Happiness Report(2016)

* Country 
* Region
* Happiness Rank
* Lower Confidence Interval
* Upper Confidence Interval
* Economy
* Family
* Health
* Freedom -> Freedom of the individual in the country 
* Trust -> Government Corruption 
* Generosity
* Dystopia Residual -> Dystopia is an imaginary country that has the worldâ€™s least-happy people. The purpose in establishing Dystopia is to have a benchmark against which all countries can be favorably compared. Dystopia Residual is caculated on the six variables(lowest incomes, lowest life expectancy, lowest generosity, most corruption, least freedom and least social support) with all the countries in the dataset.

# Importing Libraries for Exploratory Data Analysis


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train=pd.read_csv('happiness dataset.csv')
```

```python
train.head()
```

```python

train.tail ()
```

```python
train=train.rename(columns={'Happiness Rank':'Happiness_Rank','Happiness Score':'Happiness_Score'})
train.head()
```

The dataset is already sorted according to the Happiness_Rank and we understand that the top five countries are from Western Europe

# Missing Data
We can use seaborn to create a simple heatmap to check for missing data.

```python
sns.heatmap(train.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

# We conclude from the heatmap that there are no null values present in the dataset

# Counting the countries from different regions

```python
train.Region.value_counts()
```

```python
dt = train.Region.value_counts()
plt.figure(figsize=(10,7))
sns.barplot(x = dt.index, y = dt.values)
plt.xlabel('Region')
plt.xticks(rotation = 90)
plt.ylabel('Number of Countries')
plt.title('Number of Countries According to the Region', color = 'blue', fontsize = 20)
plt.show()
```

```python
x = sns.stripplot(x = "Region", y = "Happiness_Score", data =train, jitter = True)
plt.xticks(rotation = 90)
plt.title("Countries' Happiness Score According to the Region", color = 'blue', fontsize = 15)
plt.show()
```

* Western Europe Countries have the best average happiness score 
* Not enough data to conclude that southern Asia has a poor happiness score
* Most of the countries in Sub-Saharan Africa have low happiness score

# Checking Correlation

```python
f,ax = plt.subplots(figsize = (12, 12))
sns.heatmap(train.corr(), annot = True, linewidths = 0.1, fmt = '.1f', ax = ax, square = True)
```

# Hapiness Score correlation with different attributes
* Strongly correlated with the Economy
* Strongly correlated with Family
* Strongly correlated with Health(Life Expectancy)
* Average correlation with Freedom 


```python
above_mean_HS =['Above World Average' if i >= 5.35 else 'Below World Average' for i in train.Happiness_Score]
df = pd.DataFrame({'Happiness_Score' : above_mean_HS})

sns.countplot(x = df.Happiness_Score)
plt.xlabel('Happiness Score')
plt.ylabel('Number of Countries')
plt.title('Number of Countries based on Happiness Score Average', color = 'blue', fontsize = 15)
plt.show()
```

### No big difference between the number of countries which are above average and the below average
